```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
A1 = np.arange(25).reshape(5,5)
A1
```

```python
B1 = np.random.rand(25).reshape(5,5)
B1
```

```python
np.concatenate([A1,B1], axis=1)
```

```python
#concatenate on pandas

s1 = Series([100,200,300])
s2 = Series([400,500,600])
print(s1)
print(s2)
```

```python
pd.concat([s1,s2], axis=1)
```

```python
pd.concat([s1,s2], axis=0)
```

```python
#concatination in DATAFRAME

df1 = DataFrame(np.arange(25).reshape(5,5))
df1
```

```python
df2 = DataFrame(np.arange(25).reshape(5,5))
df2
```

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=0,ignore_index=True)
```

```python
pd.concat([df1,df2],axis=1)
```
